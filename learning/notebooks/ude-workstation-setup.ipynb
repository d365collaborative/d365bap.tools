# Setup your Unified Development Environment (UDE) workstation

d365bap.tools provides several commands that help you set up a Unified Development Environment (UDE) workstation. This notebook is intended as a companion guide to the official Microsoft documentation on how to [Install and configure development tools](https://learn.microsoft.com/en-us/power-platform/developer/unified-experience/finance-operations-install-config-tools).

## Visual Studio prerequisites

One of the prerequisites for setting up a Unified Development Environment (UDE) workstation is to have Visual Studio 2022 installed with the necessary components and extensions. The `Confirm-UdeVs2022Installation` command from d365bap.tools can be used to verify this and install missing components if needed. This is based on the [Full-Ude.vsconfig](https://github.com/d365collaborative/d365bap.tools/blob/main/d365bap.tools/internal/misc/Full-Ude.vsconfig) and [Vs2022.Extensions.json](https://github.com/d365collaborative/d365bap.tools/blob/main/d365bap.tools/internal/misc/Vs2022.Extensions.json) files. These files along with the downloaded extensions will be placed in `C:\Temp\d365bap.tools\Vs2022-Ude-Prerequisites` in case you need to refer to them.

> ℹ️NOTE: When running this command for the first time and no other Visual Studio extension have been installed before, it may fail with an error:<br>
> `Start-Process: This command cannot be run completely because the system cannot find all the information required.`<br>
> When running this command from a terminal, a dialog "Select an app to open this .vsix file" may appear. In this case select "Microsoft Visual Studio Version Selector" and then select the Visual Studio version you are using. Afterwards, rerun the command. Additional extension should now install without error.<br>
> If you are running this command from this notebook, you may need to run the command in an external PowerShell terminal instead.

```polyglot-notebook
Confirm-UdeVs2022Installation
```
